```python
# Loop through each sublist in grads1 and plot it using a blue line
for i in range(len(grads1)):
    plt.plot(grads1[i], 'b')

# Draw a horizontal line at y=0 across the entire plot
plt.hlines(0, 0, len(grads1[0])+1, linewidth=1, color="k" )

# Remove x-axis ticks
plt.xticks([])

# Set x-axis limits from 0 to the length of the first sublist in grads1
plt.xlim(xmin=0, xmax=len(grads1[0]))

# Disable TeX rendering for text
plt.rcParams['text.usetex'] = False

# Set x-axis label
plt.xlabel('Layers (shallower $\\rightarrow$ deeper)')

# Set y-axis label
plt.ylabel('Average gradient')

# Set plot title
plt.title('Gradient flow')

# Display grid on the plot
plt.grid(True)

# Adjust layout for better appearance
plt.tight_layout()

# Add a legend to the plot with the label 'Unfiltered'
plt.legend(['Unfiltered'])

# Get the current Axes instance
ax = plt.gca()

# Get the legend handles and set the color of the first handle to blue
leg = ax.get_legend()
leg.legendHandles[0].set_color('blue')

# Save the plot as an image file named 'grad_view'
plt.savefig("grad_view")
```
